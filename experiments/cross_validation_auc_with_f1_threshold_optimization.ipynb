```python
# ========================== BLOQUE 1 ==========================
# üìå CARGA DE DATOS Y B√öSQUEDA DE HIPERPAR√ÅMETROS CON OPTUNA USANDO AUC
import awswrangler as wr
import pandas as pd
import optuna
import mlflow
import mlflow.sklearn
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import KFold
from sklearn.metrics import auc, roc_auc_score, roc_curve
import numpy as np
import matplotlib.pyplot as plt

# üöÄ Cargar los datos desde AWS S3
X_train = wr.s3.read_csv("s3://data/final/train/X_train_scaled.csv")
y_train = wr.s3.read_csv("s3://data/final/train/y_train.csv")
X_val = wr.s3.read_csv("s3://data/final/val/X_val_scaled.csv")
y_val = wr.s3.read_csv("s3://data/final/val/y_val.csv")
X_test = wr.s3.read_csv("s3://data/final/test/X_test_scaled.csv")
y_test = wr.s3.read_csv("s3://data/final/test/y_test.csv")

# üîπ Configurar MLflow
mlflow.set_tracking_uri('http://localhost:5000') 
mlflow.set_experiment("Cross Validation AUC with F1 Threshold Optimization")

# üîπ Combinar conjuntos de entrenamiento y validaci√≥n para CV
X_combined = pd.concat([X_train, X_val])
y_combined = pd.concat([y_train, y_val])

# üìå Funci√≥n objetivo para Optuna con validaci√≥n cruzada usando AUC
def objective(trial):
    classifier_name = trial.suggest_categorical('classifier', 
        ['RandomForest', 'XGBoost', 'SVC', 'LogisticRegression', 'KNN'])

    n_splits = 5  # N√∫mero de folds para validaci√≥n cruzada
    kf = KFold(n_splits=n_splits, shuffle=True, random_state=42)

    # üîπ Definir hiperpar√°metros seg√∫n el modelo
    if classifier_name == 'RandomForest':
        model = RandomForestClassifier(
            n_estimators=trial.suggest_int('n_estimators', 10, 100),
            max_depth=trial.suggest_int('max_depth', 2, 32)
        )
    elif classifier_name == 'XGBoost':
        model = XGBClassifier(
            eta=trial.suggest_float('eta', 0.01, 0.5),
            max_depth=trial.suggest_int('max_depth', 2, 20),
            use_label_encoder=False
        )
    elif classifier_name == 'SVC':
        model = SVC(C=trial.suggest_float('C', 0.1, 10.0, log=True), probability=True)
    elif classifier_name == 'LogisticRegression':
        model = LogisticRegression(
            C=trial.suggest_float('C', 0.1, 10.0, log=True),
            penalty=trial.suggest_categorical('penalty', ['l1', 'l2']),
            solver='liblinear'
        )
    elif classifier_name == 'KNN':
        model = KNeighborsClassifier(
            n_neighbors=trial.suggest_int('n_neighbors', 1, 30),
            weights=trial.suggest_categorical('weights', ['uniform', 'distance']),
            p=trial.suggest_int('p', 1, 2)
        )

    # üîπ Validaci√≥n cruzada
    cv_auc_scores = []
    with mlflow.start_run(run_name=f"{classifier_name}_cv_run"):
        mlflow.log_param("classifier", classifier_name)
        mlflow.log_params(trial.params)

        for train_idx, val_idx in kf.split(X_combined):
            X_fold_train, y_fold_train = X_combined.iloc[train_idx], y_combined.iloc[train_idx]
            X_fold_val, y_fold_val = X_combined.iloc[val_idx], y_combined.iloc[val_idx]

            model.fit(X_fold_train, y_fold_train.values.ravel())
            y_probs = model.predict_proba(X_fold_val)[:, 1]  # Obtener probabilidades de clase positiva
            cv_auc_scores.append(roc_auc_score(y_fold_val, y_probs))

        mean_cv_auc = np.mean(cv_auc_scores)
        mlflow.log_metric("mean_cv_auc", mean_cv_auc)
        
        # üîπ Ahora simplemente usamos el modelo ya entrenado para la curva ROC sin volver a entrenar
        y_probs_test = model.predict_proba(X_test)[:, 1]  # Probabilidades en test

        # Calcular la curva ROC
        fpr, tpr, _ = roc_curve(y_test, y_probs_test)
        roc_auc = auc(fpr, tpr)

        # Crear la gr√°fica ROC
        plt.figure(figsize=(8, 6))
        plt.plot(fpr, tpr, color='blue', lw=2, label=f'ROC Curve (AUC = {roc_auc:.4f})')
        plt.plot([0, 1], [0, 1], color='gray', linestyle='--')
        plt.xlabel("False Positive Rate")
        plt.ylabel("True Positive Rate")
        plt.title(f"Curva ROC - {classifier_name}")
        plt.legend(loc="lower right")

        # Guardar la imagen de la curva ROC en MLflow sin reentrenar
        plt.savefig("roc_curve.png")
        mlflow.log_artifact("roc_curve.png")  # üìå Guardar en MLflow

        print(f"‚úÖ Curva ROC guardada en MLflow para {classifier_name}")

        return mean_cv_auc

# üöÄ Ejecutar la optimizaci√≥n con Optuna
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=50)

# üîπ Guardar el mejor modelo y sus hiperpar√°metros
best_params = study.best_trial.params
best_classifier_name = best_params.pop("classifier")

print(f"üèÜ Mejor modelo: {best_classifier_name} con par√°metros: {best_params}")

# üîπ Entrenar el mejor modelo con los hiperpar√°metros √≥ptimos
if best_classifier_name == 'RandomForest':
    best_model = RandomForestClassifier(**best_params)
elif best_classifier_name == 'XGBoost':
    best_model = XGBClassifier(**best_params, use_label_encoder=False)
elif best_classifier_name == 'SVC':
    best_model = SVC(**best_params, probability=True)
elif best_classifier_name == 'LogisticRegression':
    best_model = LogisticRegression(**best_params, solver='liblinear')
elif best_classifier_name == 'KNN':
    best_model = KNeighborsClassifier(**best_params)

best_model.fit(X_combined, y_combined.values.ravel())
```

```python
# ========================== BLOQUE 2 ==========================
# üìå OPTIMIZACI√ìN DEL UMBRAL PARA MEJORAR EL F1-SCORE MACRO
import matplotlib.pyplot as plt
from sklearn.metrics import precision_recall_curve

# üîπ Obtener probabilidades del modelo √≥ptimo en el set de prueba
y_probs = best_model.predict_proba(X_test)[:, 1]  # Solo probabilidades de la clase positiva

# üîπ Calcular precisi√≥n, recall y umbrales posibles
precisions, recalls, thresholds = precision_recall_curve(y_test, y_probs)

# üìå Asegurar que `precisions` y `recalls` tengan la misma longitud que `thresholds`
precisions = precisions[:-1]
recalls = recalls[:-1]


# üîπ Calcular F1-score para cada umbral
f1_scores = 2 * (precisions * recalls) / (precisions + recalls)
best_idx = np.argmax(f1_scores)
best_threshold = thresholds[best_idx]
best_f1 = f1_scores[best_idx]

print(f"üìå Mejor umbral encontrado: {best_threshold:.2f} con F1-score macro: {best_f1:.4f}")

# üîπ Graficar F1-score en funci√≥n del umbral
plt.figure(figsize=(8, 5))
plt.plot(thresholds, f1_scores, label="F1-score macro")
plt.axvline(best_threshold, color="r", linestyle="--", label=f"Mejor umbral: {best_threshold:.2f}")
plt.xlabel("Umbral de Clasificaci√≥n")
plt.ylabel("F1-score macro")
plt.title("Optimizaci√≥n del Umbral para F1-score Macro")
plt.legend()
plt.show()

# üîπ Hacer predicciones con el umbral ajustado
y_pred_adjusted = (y_probs >= best_threshold).astype(int)
```
